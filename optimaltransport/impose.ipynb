```python
import matplotlib.pyplot as plt
from IPython import display
import numpy as np
import cv2
from scipy.stats import gaussian_kde
import seaborn as sns

import warnings

warnings.filterwarnings('ignore')

plt.style.use('default')
plt.rc('text', usetex=True)
plt.rc('font', family='sans-serif')
plt.rc('font', size=14)
plt.rc('axes', titlesize=14)
plt.rc('axes', labelsize=14)
plt.rc('xtick', labelsize=14)
plt.rc('ytick', labelsize=14)
plt.rc('legend', fontsize=14)
plt.rc('lines', markersize=10)
```

```python
PATH_TO_IMAGE = 'fic.jpeg'
image = cv2.imread(PATH_TO_IMAGE)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
sns.kdeplot(image.ravel(), ax=ax, color='black', fill=True)
ax.set_xlabel('Pixel value')
ax.set_ylabel('Density')
ax.set_title('Density of pixels in the image')
plt.show()
```

```python
new_image = np.zeros(image.shape)
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        new_image[i, j, :] = image[i, j, :] * 3

new_image = new_image.astype(np.uint8)
new_image = new_image / 255
image = image / 255
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
sns.kdeplot(image.ravel(), ax=ax[0], color='black', fill=True)
ax[0].set_xlabel('Pixel value')
ax[0].set_ylabel('Density')
ax[0].set_title('Density of Pixels in the Image')
sns.kdeplot(new_image.ravel(), ax=ax[1], color='blue', fill=True)
ax[1].set_xlabel('Pixel value')
ax[1].set_ylabel('Density')
ax[1].set_title('Density of Pixels in the New Image')
plt.tight_layout()
plt.show()
```

```python
# Plot new image
fig, ax = plt.subplots(figsize=(5, 5))
ax.imshow(new_image)
ax.set_title('New Image')
plt.show()
```
