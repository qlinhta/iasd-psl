```python
"""def newtons_method(objective_function, gradient_function, hessian_function, starting_point,
                   convergence_tolerance=1e-6, maximum_iterations=1000, enable_verbose=False):
    global table
    optimal_point = np.array(starting_point, dtype=float)
    points_history = [optimal_point.copy()]

    if enable_verbose:
        table = PrettyTable(["Iteration", "Optimal Point", "Function Value", "Step Size", "Iteration Time"])
        table.add_row(
            [0, np.round(starting_point, 5).tolist(), round(objective_function(starting_point), 5), "N/A", "N/A"])

    for iteration_number in range(1, maximum_iterations + 1):
        start_time = time.time()

        gradient_at_current_point = gradient_function(optimal_point)
        hessian_at_current_point = hessian_function(optimal_point)

        try:
            inverse_hessian = np.linalg.inv(hessian_at_current_point)
            optimization_step = inverse_hessian @ gradient_at_current_point
            step_size = np.linalg.norm(optimization_step)
            next_point = optimal_point - optimization_step
            points_history.append(next_point.copy())

            end_time = time.time()
            duration = round(end_time - start_time, 5)

            if step_size < convergence_tolerance:
                final_value = round(objective_function(next_point), 5)
                if enable_verbose:
                    if iteration_number == 1 or step_size > 0:
                        table.add_row(
                            [iteration_number, np.round(next_point, 5).tolist(), final_value, round(step_size, 5),
                             f"{duration:.5f} seconds"])
                    print(table)
                    print(f"Convergence achieved in {iteration_number} iterations.")
                break

            if enable_verbose and iteration_number == 1:
                current_value = round(objective_function(next_point), 5)
                table.add_row([iteration_number, np.round(next_point, 5).tolist(), current_value, round(step_size, 5),
                               f"{duration:.5f} seconds"])

            optimal_point = next_point

        except np.linalg.LinAlgError:
            if enable_verbose:
                print(f"Failed to invert Hessian matrix at iteration {iteration_number}. Optimization aborted.")
            break

    return optimal_point, points_history"""


def newtons_method(objective_function, gradient_function, hessian_function, starting_point,
                   convergence_tolerance=1e-10, maximum_iterations=1000, enable_verbose=False):
    global iteration_number
    optimal_point = np.array(starting_point, dtype=float)
    points_history = [optimal_point.copy()]

    table = PrettyTable(["Iteration", "Optimal Point", "Function Value", "Step Size", "Iteration Time"])
    initial_f_val = objective_function(optimal_point)
    table.add_row([0, [round(x, 5) for x in optimal_point], round(initial_f_val, 5), "N/A", "N/A"])

    for iteration_number in range(1, maximum_iterations + 1):
        start_time = time.time()

        gradient_at_current_point = gradient_function(optimal_point)
        hessian_at_current_point = hessian_function(optimal_point)

        try:
            inverse_hessian = np.linalg.inv(hessian_at_current_point)
            optimization_step = inverse_hessian @ gradient_at_current_point
            next_point = optimal_point - optimization_step
            step_size = np.linalg.norm(optimization_step)

            if step_size < convergence_tolerance or np.allclose(optimal_point, next_point, atol=convergence_tolerance):
                if iteration_number == 1:
                    duration = round(time.time() - start_time, 5)
                    next_f_val = objective_function(next_point)
                    table.add_row(
                        [iteration_number, [round(x, 5) for x in next_point], round(next_f_val, 5), round(step_size, 5),
                         f"{duration:.5f} seconds"])
                break

            points_history.append(next_point)
            optimal_point = next_point

            if enable_verbose:
                duration = round(time.time() - start_time, 5)
                next_f_val = objective_function(next_point)
                table.add_row(
                    [iteration_number, [round(x, 5) for x in next_point], round(next_f_val, 5), round(step_size, 5),
                     f"{duration:.5f} seconds"])

        except np.linalg.LinAlgError:
            print("Failed to invert Hessian matrix. Optimization aborted.")
            break

    if enable_verbose:
        print(table)
        print(f"Convergence achieved in {iteration_number - 1} iterations.")

    return optimal_point, points_history
```

### Solution 1.1

Given the quadratic optimization problem:
$$ q(\mathbf{w}) = 2(w_1 + w_2 + w_3 - 3)^2 + (w_1 - w_2)^2 + (w_2 - w_3)^2 $$
We define the objective function $f(\mathbf{w})$ and aim to minimize it using Newton's method. The gradient $\nabla f(\mathbf{w})$ and the Hessian $\nabla^2 f(\mathbf{w})$ are essential in this process.

First, let's establish the gradient and Hessian of the function:

The gradient of $f(\mathbf{w})$ is given by:
$$
\nabla f \begin{pmatrix}
w_1 \\
w_2 \\
w_3
\end{pmatrix} = \begin{pmatrix}
6w_1 +2w_2 +4w_3 -12 \\
2w_1 +8w_2 +2w_3 -12 \\
4w_1 + 2w_2 +6w_3 -12 
\end{pmatrix}
$$

The Hessian of $f(\mathbf{w})$, and its inverse, are given by:
$$
\nabla^2 f \begin{pmatrix}
w_1 \\
w_2 \\
w_3
\end{pmatrix} = \begin{pmatrix}
6 & 2 & 4 \\
2 & 8 & 2 \\
4 & 2 & 6 
\end{pmatrix}, \quad \left(\nabla^2 f \begin{pmatrix}
w_1 \\
w_2 \\
w_3
\end{pmatrix}\right)^{-1} = \frac{1}{36} \begin{pmatrix}
11 & -1 & -7 \\
-1 & 5 & -1 \\
-7 & -1 & 11 
\end{pmatrix}
$$

Applying Newton's method, we update $\mathbf{w}$ as follows:
$$
\mathbf{w} = \mathbf{w}_0 - [\nabla^2 f(\mathbf{w}_0)]^{-1} \nabla f(\mathbf{w}_0)
$$

By substituting the expressions for $\nabla f(\mathbf{w}_0)$ and $[\nabla^2 f(\mathbf{w}_0)]^{-1}$, we obtain:
$$
\mathbf{w} = \begin{pmatrix}
w_1 \\
w_2 \\
w_3
\end{pmatrix} - \frac{1}{36} \begin{pmatrix}
11 & -1 & -7 \\
-1 & 5 & -1 \\
-7 & -1 & 11 
\end{pmatrix} \times \begin{pmatrix}
6w_1 +2w_2 +4w_3 -12 \\
2w_1 +8w_2 +2w_3 -12 \\
4w_1 + 2w_2 +6w_3 -12 
\end{pmatrix}
$$

This simplifies to:
$$
\mathbf{w} = \begin{pmatrix}
w_1 \\
w_2 \\
w_3
\end{pmatrix} - \begin{pmatrix}
w_1-1 \\
w_2-1 \\
w_3-1
\end{pmatrix} = \begin{pmatrix}
1 \\
1 \\
1
\end{pmatrix}
$$

Hence, the algorithm converges to the solution $\mathbf{w} = \begin{pmatrix} 1 \\ 1 \\ 1 \end{pmatrix}$ in just one iteration. To verify the optimality of this solution, we evaluate the objective function at $\mathbf{w}$:
$$
f(\mathbf{w}) = 2(1+1+1-3)^2 + (1-1)^2 + (1-1)^2 = 0
$$

This result confirms that $\mathbf{w} = \begin{pmatrix} 1 \\ 1 \\ 1 \end{pmatrix}$ is indeed the minimizer of $f$, as the function value at this point is the global minimum.

In conclusion, this proof demonstrates that Newton's method, when applied to the given quadratic optimization problem, converges to the global minimum in a single iteration, showcasing the efficiency and power of Newton's method for this class of problems.

```python
def function(w):
    return 2 * (w[0] + w[1] + w[2] - 3) ** 2 + (w[0] - w[1]) ** 2 + (w[1] - w[2]) ** 2


def grad(w):
    return np.array(
        [6 * w[0] + 2 * w[1] + 4 * w[2] - 12, 2 * w[0] + 8 * w[1] + 2 * w[2] - 12, 4 * w[0] + 2 * w[1] + 6 * w[2] - 12])


def hess(w):
    return np.array([[6, 2, 4], [2, 8, 2], [4, 2, 6]])
```

```python
initial_points = [np.array([0, 0, 0]), np.array([-1, 2, -1]), np.array([2, -1, 3])]
results = [newtons_method(function, grad, hess, point, enable_verbose=True) for point in
           initial_points]
```

```python
def rosenbrock(w):
    return 100 * (w[1] - w[0] ** 2) ** 2 + (1 - w[0]) ** 2


def grad_rosenbrock(w):
    return np.array([
        -400 * w[0] * (w[1] - w[0] ** 2) - 2 * (1 - w[0]),
        200 * (w[1] - w[0] ** 2)
    ])


def hess_rosenbrock(w):
    return np.array([
        [800 * w[0] ** 2 - 400 * (w[1] - w[0] ** 2) + 2, -400 * w[0]],
        [-400 * w[0], 200]
    ])
```

```python
starting_points = [
    np.array([-1.2, 1]),
    np.array([0, 1 / 200 + 1e-12])
]

results_rosenbrock = [newtons_method(rosenbrock, grad_rosenbrock, hess_rosenbrock, point, enable_verbose=True) for point
                      in
                      starting_points]
```
